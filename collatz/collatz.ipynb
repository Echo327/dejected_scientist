```python
n = 50
```

```python
print(n)
while n != 1:
    if n%2 == 0:
        n /= 2
    elif n%2 == 1:
        n = 3*n + 1
    print(int(n))
```

```python
def count_collatz(n):
    count = 0
    while n != 1:
        if n%2 == 0:
            n /= 2
        elif n%2 == 1:
            n = 3*n + 1
        count += 1
    return count
```

```python
n1 = [2**i for i in range(1,101)]
print(n1)
```

```python
counter = [count_collatz(i) for i in n1]
print(counter)
```
